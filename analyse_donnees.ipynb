```python
import numpy as np
import matplotlib.pyplot as plt
# On importe les bibliothèques nécessaires
# numpy pour générer des données aléatoires
# matplotlib pour les afficher sous forme de graphique

data = np.random.randn(1000)
# Génère 1000 nombres aléatoires suivant une loi normale (moyenne=0, écart-type=1)

plt.hist(data, bins=30, alpha=0.75)
# Trace un histogramme avec 30 barres, légèrement transparent (alpha=0.75)

plt.title('Distribution des données')
# Ajoute un titre au graphique

plt.show()
# Affiche le graphique

plt.boxplot(data)
plt.title('Boîte à moustaches')
plt.show()
# On ajoute une deuxième visualisation dans le notebook
# Cette modification existe uniquement sur la branche "nouvelle-fonctionnalite"
```
